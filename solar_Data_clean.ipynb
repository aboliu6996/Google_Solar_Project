```python
import pandas as pd
```

```python
df = pd.read_csv("project-sunroof-postal_code-09082017.csv")
df.head()
```

```python
df.info()
```

```python
# List of columns to drop
columns_to_drop = [
    "yearly_sunlight_kwh_median",
    "yearly_sunlight_kwh_total",
    "install_size_kw_buckets_json",
    "carbon_offset_metric_tons",
    "carbon_offset_metric_tons",
    "number_of_panels_median",
    "number_of_panels_total",
    "kw_median",
    "kw_total"
]

# Remove leading and trailing whitespaces from column names
columns_to_drop_stripped = [column.strip() for column in columns_to_drop]

# Drop columns from the DataFrame
clean_zip_data = df.drop(columns=columns_to_drop_stripped)
clean_zip_data.head()
```

```python
clean_zip_data.info()
```

```python
clean_zip_data.drop(index=0, inplace=True)
clean_zip_data.info()
```

```python
# Save the DataFrame to a CSV file
clean_zip_data.to_csv('clean_zip_data.csv', index=False)
```
